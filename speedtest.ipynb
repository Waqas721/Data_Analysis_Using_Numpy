```python
#using python list
import time 
size=1_000_000_0
l1=list(range(size))
l2=list(range(size))
start=time.time()
add=[x+y for x,y in zip(l1,l2)]
end=time.time()
print(add[0:20])
print(end-start)
```

```python
#using numpy arrays
import numpy as np 
size=1_000_000_0
l1=np.array(list(range(size)))
l2=np.array(list(range(size)))
start=time.time()
add=l1+l2
end=time.time()
print(add[0:20])
print(end-start)
```

```python
#Compare the time taken to multiply two Python lists vs. two NumPy arrays.
list1=[x for x in range(1,6)]
list2=[x for x in range(6,11)]
start=time.time()
mul_list=[x*y for x,y in zip(list1,list2)]
end=time.time()
print(mul_list,"time:",end-start)
arr1=np.array(list1)
arr2=np.array(list2)
start1=time.time()
mul_arr=arr1*arr2
end1=time.time()
print(mul_arr,"time:",end1-start1)
```
